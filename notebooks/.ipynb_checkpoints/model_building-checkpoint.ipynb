Build a regression model.

```python
#hypothesis is there is relationship between the locations(IV) and the number of bikes(DV) in the nearest station
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

df_final_resul = pd.read_csv('result.csv') 

# Create a feature DataFrame
features = df_final_resul[['latitude', 'longitude']]

# Add a constant to the features, as statsmodels' OLS function does not include it by default
features = sm.add_constant(features)

# Create a target DataFrame
target = df_final_resul['num_bikes']

# Create a OLS model
model = sm.OLS(target, features)

# Fit the model
results = model.fit()

# Print the summary statistics of the regression model
print(results.summary())


```

```python
# To visualize the results, we can create a 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# The actual values
ax.scatter(df_final_resul['latitude'], df_final_resul['longitude'], df_final_resul['num_bikes'], color='blue', label='Actual Value')

# The predicted values
ax.scatter(df_final_resul['latitude'], df_final_resul['longitude'], results.fittedvalues, color='red', label='Predicted Value')

# Setting labels
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
ax.set_zlabel('Num_bikes')

# Setting the legend
plt.legend()

# Showing the plot
plt.show()
```

Provide model output and an interpretation of the results. 

```python
#p value is ≤ 0.05 so we reject the null hypothesis
#the log-likehood indicating that the data is quite unlikely
#R-square only 21.8% of the total variance in the dependent variable can be explained by the independent variable
#my model need more features 
```
