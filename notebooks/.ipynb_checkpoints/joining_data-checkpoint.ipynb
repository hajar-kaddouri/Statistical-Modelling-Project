Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
import pandas as pd
from sklearn.neighbors import BallTree

# Load the CSV files 
df_stations = pd.read_csv('citybike.csv')
df_restaurants = pd.read_csv('yelp-foursquare.csv')

# Build a ball tree on df_stations
tree = BallTree(df_stations[['latitude', 'longitude']], leaf_size=2)

# Query the ball tree to find the index of the nearest neighbor in df_stations for each point in df_restaurants
dist, ind = tree.query(df_restaurants[['latitude', 'longitude']], k=1)

# Map the index to the corresponding station_name in df_stations
df_restaurants['nearest_station'] = df_stations.loc[ind.flatten(), 'station_name'].values

# Map the index to the corresponding num_bikes in df_stations
df_restaurants['num_bikes'] = df_stations.loc[ind.flatten(), 'num_bikes'].values

# Now, each restaurant in df_restaurants has a 'nearest_station' column with the name of the nearest station and number of bike existing
print(df_restaurants)

# Write the result to a new CSV file
df_restaurants.to_csv('result.csv',index=False)
```

Provide a visualization that you used as part of your EDA process. Explain the initial pattern or relationship you discoved through this visualization. 

```python
import pandas as pd 

import matplotlib.pyplot as plt
import seaborn as sns

df_final_resul = pd.read_csv('result.csv') 

# Creating a scatter plot
plt.figure(figsize=(10,6))
sns.scatterplot(x='nearest_station', y='name', hue='distance', size='distance', sizes=(20, 200), palette='viridis', data=df_final_resul)

# Setting plot title and labels
plt.title('Restaurants and Their Nearest Stations')
plt.xlabel('stations')
plt.ylabel('restaurants')

# Displaying the plot
plt.show()

```

# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
import sqlite3
import pandas as pd

# Load the data
df_final_resul = pd.read_csv('result.csv')  

# Connect to the SQLite database 
conn = sqlite3.connect('citybike_database.db')

# Write the data to a new table in the database
df_final_resul.to_sql('citybike', conn, if_exists='replace', index=False)

# Close the connection
conn.close()
```
